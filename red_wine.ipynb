We can find correlations using pandas “.corr()” function and can visualize the correlation matrix using a heatmap in seaborn.

```python
#Check the strength of the correlation among the variables.
plt.figure(figsize=(12,8))
sns.heatmap(red_wine.corr(), cmap='Greens', annot=True)
plt.show()
```

1.density has a positive correlation with fixed acidity, whereas it has a negative correlation with alcohol
2.pH & fixed acidity has negative correlation
3.citric acid & fixed acidity has positive correlation
4.citric acid & volatile acidity has negative correlation
5.free sulphur dioxide & total sulphur dioxide has positive correlation
6.Wine quality is negatively correlated with the volatile acidity, as too high levels of it leads to vinegary taste

```python
#plotting the relationship among the important variables
red_wine.corr()[['quality']].sort_values(by='quality', ascending = False)

# alcohol,volatile acidity,sulphates and citric acid influence the quality of red wine in  order. 
# the relationship among the important variables
```

Conclusion
As alcohol level increase ==> Quality increases
As the volatile acidity decreases ==> Quality increases
As sulphate increases ==> Quality increases
As citric acid level increase ==> Quality increases

```python
# We can use either PairPlot or PairGrid to perform below visualization. I prefer PairGrid to have better control of my figure.
g = sns.PairGrid(red_wine[['alcohol','volatile acidity','sulphates','citric acid']])
g = g.map_upper(sns.kdeplot, cmap='YlGnBu', shade=True, thresh=0.05)
g = g.map_diag(plt.hist, color='#6cbd5e')
g = g.map_lower(sns.scatterplot, color='#b8e0b1', edgecolor='k')

g.fig.tight_layout()
g.fig.suptitle('Relationship')
plt.subplots_adjust(top=0.9)
```

```python
# Plot JointPlot
#1. fixed acidity~pH(-0,68)
g = sns.JointGrid(x='fixed acidity', y='pH', data=red_wine)
g = g.plot_joint(sns.regplot, color='#eb447e')
```

```python
# Plot JointPlot
# 1.fixed acidity~density(0,67)

g = sns.JointGrid(x='fixed acidity', y='density', data=red_wine)
g = g.plot_joint(sns.regplot, color='#eb447e')
#g = g.plot_marginals(sns.distplot, color='#fc7f03')
```

```python
# Plot JointPlot
# 3.fixed acidity~citric acid(0,67)

g = sns.JointGrid(x='fixed acidity', y='citric acid', data=red_wine)
g = g.plot_joint(sns.regplot, color='#eb447e')
#g = g.plot_marginals(sns.distplot, color='#fc7f03')
```

```python
# Plot JointPlot
# 5.free sulfur dioxide ~ total sulfur dioxide(0,67)

g = sns.JointGrid(x='free sulfur dioxide', y='total sulfur dioxide', data=red_wine)
g = g.plot_joint(sns.regplot, color='#eb447e')
#g = g.plot_marginals(sns.distplot, color='#fc7f03')
```

##  This is the box plot of various features ###
In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.

```python
plt.figure(figsize=(10,15))
for i,col in enumerate(list(red_wine.columns.values)):
    plt.subplot(4,3,i+1)
    red_wine.boxplot(col)
    plt.grid()
    plt.tight_layout()
```

seaborn.distplot- This function plots the estimated PDF (parametric distribution of a dataset) over the data.


```python
plt.figure(figsize=(20,16))
for i,col in enumerate(list(red_wine.columns.values)):
    plt.subplot(4,3,i+1)
    sns.histplot(red_wine[col],color='r',kde=True,label='data')
    plt.grid()
    plt.legend(loc='upper right')
    plt.tight_layout()
```

1. pH feature is approximately normally distributed
2. mot of independent features are positively skewed

###   About Box Plots  ###
The categorical column name is passed to the x variable while the numeric column name should be passed to the y variable. 

```python
sns.boxplot(x='quality_label', y='alcohol', data=red_wine)
```

```python
sns.boxplot(x='quality', y='alcohol', data=red_wine, hue=('quality_label'))
```

alcohol tends to be higher for the higher quality wines 6,7,8

```python
sns.boxplot(x='quality', y='pH', data=red_wine,hue=('quality_label'))
```

For pH, one can see the lower quality (3 and 4) wines generally have higher pH than the higher quality wines (7 and 8). As mentioned earlier, the almost zero correlation coefficient did not indicate this behavior.


```python
sns.boxplot(x='quality', y='volatile acidity', data=red_wine,hue=('quality_label'))
```

as wine quality increases volatile acidity decreases in general.
