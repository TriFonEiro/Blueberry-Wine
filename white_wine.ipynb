Let's explore different statistical measures that we have got from desribe().

count: total count of non-null values in the column
mean: the average of all the values in that column
min: the minimum value in the column
max: the maximum value in the column
25%: first quartile in the column after we arrange those values in ascending order
50%: this is the median or the second quartile
75%: the third quartile
std: this is the standard deviation (i.e. measure of depreciation, you must have read in the basics of statistics study material)
Note: 25%, 50%, and 75% are nothing but corresponding percentile values

We can find correlations using pandas “.corr()” function and can visualize the correlation matrix using a heatmap in seaborn.

```python
plt.figure(figsize=(12,8))
sns.heatmap(white_wine.corr(), cmap='Greens', annot=True)
plt.show()
```

1.density has a strong positive correlation with residual sugar and a moderate with total sulfur dioxide, whereas it has a negative correlation with alcohol
2.pH & fixed acidity has negative correlation
3.citric acid & fixed acidity has positive correlation
4.redidual sugar positive with tsd and negative with alcohol
5.free sulphur dioxide & total sulphur dioxide has positive correlation
6.Wine quality is negatively correlated with the density and positive with alcohol

```python
white_wine.corr()[['quality']].sort_values(by='quality', ascending = False)

# alcohol,volatile acidity,sulphates and citric acid influence the quality of red wine in  order. 
# the relationship among the important variables
```

Conclusion
As alcohol level increase ==> Quality increases
As the density decrease ==> Quality increases
As chlorides decrease ==> Quality increases
As volatile acidity decrease ==> Quality increases

```python
g = sns.PairGrid(white_wine[['alcohol','density','chlorides','volatile acidity']])
g = g.map_upper(sns.kdeplot, cmap='YlGnBu', shade=True, thresh=0.05)
g = g.map_diag(plt.hist, color='#6cbd5e')
g = g.map_lower(sns.scatterplot, color='#b8e0b1', edgecolor='k')

g.fig.tight_layout()
g.fig.suptitle('Relationship')
plt.subplots_adjust(top=0.9)
```

```python
# Plot JointPlot
#1. density~alcohol(-0,78)
g = sns.JointGrid(x='alcohol', y='density', data=white_wine)
g = g.plot_joint(sns.regplot, color='#eb447e')
```

```python
# Plot JointPlot
# 1.density~residual sugar(0,84)

g = sns.JointGrid(x='residual sugar', y='density', data=white_wine)
g = g.plot_joint(sns.regplot, color='#eb447e')
#g = g.plot_marginals(sns.distplot, color='#fc7f03')
```

```python
# Plot JointPlot
# 5.free sulfur dioxide ~ total sulfur dioxide(0,62)

g = sns.JointGrid(x='free sulfur dioxide', y='total sulfur dioxide', data=white_wine)
g = g.plot_joint(sns.regplot, color='#eb447e')
#g = g.plot_marginals(sns.distplot, color='#fc7f03')
```

##  This is the box plot of various features ###
In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.

```python
plt.figure(figsize=(10,15))
for i,col in enumerate(list(white_wine.columns.values)):
    plt.subplot(4,3,i+1)
    white_wine.boxplot(col)
    plt.grid()
    plt.tight_layout()
```

seaborn.distplot- This function plots the estimated PDF (parametric distribution of a dataset) over the data.


```python
plt.figure(figsize=(20,16))
for i,col in enumerate(list(white_wine.columns.values)):
    plt.subplot(4,3,i+1)
    sns.histplot(white_wine[col],color='r',kde=True,label='data')
    plt.grid()
    plt.legend(loc='upper right')
    plt.tight_layout()
```

1. pH feature is approximately normally distributed
2. mot of independent features are positively skewed
